#### Social Score of Brands

Importing libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
```

Reading files

```python
week1=pd.read_csv('Market/Social Score Assignment - Dec_week1.csv')
week2=pd.read_csv('Market/Social Score Assignment - Dec_week2.csv')
week3=pd.read_csv('Market/Social Score Assignment - Dec_week3.csv')
week4=pd.read_csv('Market/Social Score Assignment - Dec_week4.csv')
```

Inserting new column in each dataset and merging:

```python
week1['week_no.']=1
week1 = week1.reindex(['week_no.','brand','post_id','description','commentCount','dislike','likes','retweet','shareCount','videoDuration','views','Channel'], axis=1)
week1.head()
```

```python
week2['week_no.']=2
week2 = week2.reindex(['week_no.','brand','post_id','description','commentCount','dislike','likes','retweet','shareCount','videoDuration','views','Channel'], axis=1)
week2.head()
```

```python
week3['week_no.']=3
week3 = week3.reindex(['week_no.','brand','post_id','description','commentCount','dislike','likes','retweet','shareCount','videoDuration','views','Channel'], axis=1)
week3.head()
```

```python
week4['week_no.']=4
week4 = week4.reindex(['week_no.','brand','post_id','description','commentCount','dislike','likes','retweet','shareCount','videoDuration','views','Channel'], axis=1)
week4.head()
```

Now combining to make one dataset :

```python
month_df=pd.concat([week1, week2,week3,week4])
month_df.head()
```

#### Twitter data

```python
twitter=['week_no.','brand','likes','retweet','videoDuration','Channel']
df_twitter=month_df[twitter][month_df[twitter]['Channel']=='twitter']
df_twitter.reset_index(drop=True,inplace=True)
df_twitter.head()
```

```python
df_twitter.info()
```

```python
Columns=['likes','videoDuration']
for i in Columns:
    df_twitter[i]=df_twitter[i].str.replace(',','')
    df_twitter[i]=df_twitter[i].astype('int64')
```
