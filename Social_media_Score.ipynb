#### Social Score of Brands

Importing libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
```

Reading files

```python
week1=pd.read_csv('Market/Social Score Assignment - Dec_week1.csv')
```

```python
week1.head()
```

```python
week2=pd.read_csv('Market/Social Score Assignment - Dec_week2.csv')
week2.head()
```

```python
week3=pd.read_csv('Market/Social Score Assignment - Dec_week3.csv')
week3.head()
```

```python
week4=pd.read_csv('Market/Social Score Assignment - Dec_week4.csv')
week4.head()
```

Inserting new column in each dataset and merging:

```python
week1['week_no.']=1
week1 = week1.reindex(['week_no.','brand','post_id','description','commentCount','dislike','likes','retweet','shareCount','videoDuration','views','Channel'], axis=1)
week1.head()
```

```python
week2['week_no.']=2
week2 = week2.reindex(['week_no.','brand','post_id','description','commentCount','dislike','likes','retweet','shareCount','videoDuration','views','Channel'], axis=1)
week2.head()
```

```python
week3['week_no.']=3
week3 = week3.reindex(['week_no.','brand','post_id','description','commentCount','dislike','likes','retweet','shareCount','videoDuration','views','Channel'], axis=1)
week3.head()
```

```python
week4['week_no.']=4
week4 = week4.reindex(['week_no.','brand','post_id','description','commentCount','dislike','likes','retweet','shareCount','videoDuration','views','Channel'], axis=1)
week4.head()
```

Now combining to make one dataset :

```python
month_df=pd.concat([week1, week2,week3,week4])
month_df.head()
```

##### Data Wrangling

Calculating percentage of null values:

```python
for i in month_df.columns:
    print('% of null records in feature',i,': {:.2f}%'.format(month_df[i].isnull().sum()*100/len(month_df[i])))
```

We can see from above analysis that some of the columns have more than 50% of missing data so its better we drop those particular columns along with post_id which does not give any useful info.

```python
month_df.drop(['description','dislike','shareCount','retweet','post_id'],axis=1,inplace=True)
month_df.shape
```

```python
month_df.head()
```

```python
month_df.info()
```

```python
month_df['commentCount'].nunique()
```

```python
month_df['commentCount'].value_counts()
```

```python
month_df['videoDuration'].nunique()
```

```python
month_df['videoDuration'].value_counts()
```

```python
Columns=['commentCount','likes','videoDuration','views']
for i in Columns:
    month_df[i]=month_df[i].str.replace(',','')
    month_df[i]=month_df[i].astype('float')
```

```python
month_df.info()
```

```python
month_df.describe().T
```

```python
sns.boxplot(x='views',data=month_df)
```

```python
month_df['views'].fillna(month_df['views'].median(),inplace=True)
```

```python
sns.boxplot(x='commentCount',data=month_df)
```

```python
month_df['commentCount'].fillna(month_df['commentCount'].median(),inplace=True)
```

```python
sns.boxplot(x='videoDuration',data=month_df)
```

```python
month_df['videoDuration'].fillna(month_df['videoDuration'].median(),inplace=True)
```

Converting the feature coloumn Channel into categorical feature in order to see correlation amongst all features.

```python
le=LabelEncoder()
month_df['Channel']=le.fit_transform(month_df['Channel'])
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(month_df.corr(),annot=True)
```

#### From above heatmap it is clearly visible that the features commentCount and likes have a positive correlation with each other,videoDuration and Channel have the highest positive correlation followed by views and Channel. We can also see that there is a negative correlation between likes and channel followed by commentCount and Channel.

#### We can positively conclude from above the that Lorealparis has highest social media score followed by Dove and lastly Tresemme.
